```python
import pandas as pd
import plotly.express as px

# Load the dataset
df = pd.read_csv('/ai.csv.csv')

# Assuming the CSV has columns suitable for a pie chart (e.g., 'category' and 'value')
# Replace 'category' and 'value' with your actual column names
# If your data needs transformation to fit a pie chart format, please specify.
if len(df.columns) >= 2:
    fig = px.pie(df, values=df.columns[1], names=df.columns[0], title='Pie Chart')
    fig.show()
else:
    print("The CSV does not have enough columns to generate a pie chart directly. Please specify which columns to use.")
```

```python
# Student Level Distribution
student_level_counts = df['StudentLevel'].value_counts().reset_index()
student_level_counts.columns = ['StudentLevel', 'Count']
fig_student_level = px.pie(student_level_counts, values='Count', names='StudentLevel', title='Student Level Distribution')
fig_student_level.show()
```

```python
# Discipline Distribution
discipline_counts = df['Discipline'].value_counts().reset_index()
discipline_counts.columns = ['Discipline', 'Count']
fig_discipline = px.pie(discipline_counts, values='Count', names='Discipline', title='Discipline Distribution')
fig_discipline.show()
```

```python
# Task Type Distribution
task_type_counts = df['TaskType'].value_counts().reset_index()
task_type_counts.columns = ['TaskType', 'Count']
fig_task_type = px.pie(task_type_counts, values='Count', names='TaskType', title='Task Type Distribution')
fig_task_type.show()
```

```python
# AI Assistance Level Distribution
ai_assistance_counts = df['AI_AssistanceLevel'].value_counts().reset_index()
ai_assistance_counts.columns = ['AI_AssistanceLevel', 'Count']
fig_ai_assistance = px.pie(ai_assistance_counts, values='Count', names='AI_AssistanceLevel', title='AI Assistance Level Distribution')
fig_ai_assistance.show()
```

```python
# Final Outcome Distribution
final_outcome_counts = df['FinalOutcome'].value_counts().reset_index()
final_outcome_counts.columns = ['FinalOutcome', 'Count']
fig_final_outcome = px.pie(final_outcome_counts, values='Count', names='FinalOutcome', title='Final Outcome Distribution')
fig_final_outcome.show()
```

```python
# Used Again Distribution
used_again_counts = df['UsedAgain'].value_counts().reset_index()
used_again_counts.columns = ['UsedAgain', 'Count']
fig_used_again = px.pie(used_again_counts, values='Count', names='UsedAgain', title='Used Again Distribution')
fig_used_again.show()
```
